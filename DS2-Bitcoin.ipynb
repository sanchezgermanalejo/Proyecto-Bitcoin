# Abstract

Abstract:
Este script utiliza la librería yfinance para descargar datos históricos de Bitcoin (BTC-USD) desde el 17 de septiembre de 2014 hasta el 1 de enero de 2025. Con la ayuda de pandas, se carga y visualiza una muestra de estos datos. La finalidad principal es disponer de un conjunto de datos actualizado que sirva de base para la predicción del mercado y la implementación de estrategias de trading algorítmico en Bitcoin. Finalmente, el dataset se exporta a un archivo CSV, facilitando su posterior análisis y procesamiento en proyectos de Data Science aplicados al mercado de criptomonedas.



# Importación y Carga de Datos

```python
# Instalar yfinance si no lo tienes instalado
!pip install yfinance --quiet

import yfinance as yf
import pandas as pd
from google.colab import files  # Solo si estás usando Colab

# Descargar datos históricos de Bitcoin (BTC-USD)
btc_df = yf.download("BTC-USD", start="2014-09-17", end="2024-12-31")

# Mostrar las primeras filas para verificar
print(btc_df.head())

# Guardar los datos en un archivo CSV
archivo = "btc_historical.csv"
btc_df.to_csv(archivo)
print(f"Datos guardados en {archivo}")

# Si usas Google Colab, puedes descargar el archivo a tu computadora con:
files.download(archivo)
```
